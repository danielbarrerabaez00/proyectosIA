```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
```

```python
# Leemos el csv y lo cargamos en un dataset
data = pd.read_csv('AB_NYC_2019.csv',sep=',')
```

```python
# Lo mostramos
data
```

```python
# Seleccionamos unicamente los datos a partir de los cuales podemos sacar informacion
data = data[['neighbourhood_group','neighbourhood','room_type','price','minimum_nights','number_of_reviews','reviews_per_month','calculated_host_listings_count','availability_365','week_price','PETS Allowed','swimming pool']]
```

```python
# Vemos el numero de campos nulos
print("Datos totales:",len(data),'\n')
print("Datos nulos por columna:")
print(data.isnull().sum(),'\n')
print("El numero de datos nulos supera el 25%?:")
print(data.isnull().sum()>=len(data)*0.25)

# Podemos comprobar que en ninguna de las columnas tenemos mas del 25% de
# los datos nulos, con lo que no es recomendable borrar las columnas que tenga datos nulo.
```

```python
# Vamos a optar por rellenar los datos nulos con la media del resto de valores
reviews_month_mean = round(data['reviews_per_month'].mean(),2)

data['reviews_per_month'] = data['reviews_per_month'].fillna(reviews_month_mean)

data
```

```python
le = LabelEncoder()

# Definimos un metodo que va a resumir los posibles valores de la columna de mascotas en si o no
def pets_allowed(value):
    if value in ['pets allowed', 'yes', 'we love pets']:
        return True
    else:
        return False

# Le aplicamos dicho metodo a cada fila de la columna
data['PETS Allowed'] = data['PETS Allowed'].apply(pets_allowed)

# Le hacemos un fit transform a las columnas con mas de un posible valor
data['neighbourhood_coded'] = le.fit_transform(data['neighbourhood'])
data['neighbourhood_group_coded'] = le.fit_transform(data['neighbourhood_group'])
data['room_type_coded'] = le.fit_transform(data['room_type'])

# Mostramos los datos transformados
data
```

```python
# Vemos que hay campos que no son numeros asi que hay que convertirlos
data.info()

# Convertimos los campos a numericos. Los errores se cambian por Nan
data['week_price'] = pd.to_numeric(data['week_price'], errors='coerce')
data['swimming pool'] = pd.to_numeric(data['swimming pool'], errors='coerce')
data['price'] = pd.to_numeric(data['price'], errors='coerce')

# Eliminamos las filas con NaN
data = data.dropna(subset=['week_price'])
data = data.dropna(subset=['swimming pool'])
data = data.dropna(subset=['price'])

# Convertir el tipo de datos a numericos
data['week_price'] = data['week_price'].astype(int)
data['swimming pool'] = data['swimming pool'].astype(bool)
data['price'] = data['price'].astype(float)

data.info()
```
